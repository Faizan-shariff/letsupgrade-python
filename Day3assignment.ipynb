```
num=int(input("Enter the number : "))
sum=0
while num>0:
    sum+=num
    num-=1
    print(num,"-->",sum)
print("the final sum is ",sum)
```

```
 
num = int(input("Enter any number : "))
 
if num > 1: 
 
   for i in range(2, num):
 
       if (num % i) == 0: 
 
           print(num, "is not a prime number") 
 
           break
   else: 
 
       print(num, "is a prime number") 
 
else: 
 
   print(num, "is not a prime number")
```
